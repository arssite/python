```python
import psutil
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Function to collect CPU and memory usage data
def collect_data():
    cpu_percent = psutil.cpu_percent(interval=1)
    memory_percent = psutil.virtual_memory().percent
    return cpu_percent, memory_percent
```

```python
# Function to update the plot with new data
def update_plot(frame):
    cpu_percent, memory_percent = collect_data()
    x.append(frame)
    cpu_data.append(cpu_percent)
    memory_data.append(memory_percent)
    line1.set_data(x, cpu_data)
    line2.set_data(x, memory_data)
    ax.relim()
    ax.autoscale_view()
    return line1, line2
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('Time (s)')
ax.set_ylabel('Usage (%)')
ax.set_title('CPU and Memory Usage')
line1, = ax.plot([], [], label='CPU Usage')
line2, = ax.plot([], [], label='Memory Usage')
ax.legend()
# Animate the plot with real-time data
#ani = FuncAnimation(fig, update_plot, frames=None, blit=True, interval=1000)
ani = FuncAnimation(fig, update_plot, frames=None, blit=True, interval=1000, cache_frame_data=False)
plt.show()
```
